```python
keep = ['oracle_id', 'name', 'released_at', 'layout', 'mana_cost', 'cmc', 'type_line', 'oracle_text', 'colors', 'color_identity', 'keywords', 'games', 'reserved', 'foil', 'nonfoil', 'set', 'set_name', 'set_type', 'rarity', 'flavor_text', 'artist', 'border_color', 'frame', 'power', 'toughness', 'loyalty']
for item in df.columns:
    if item not in keep:
        df = df.drop(item, axis=1)
df
```

```python
df["color_identity"] = df.apply(lambda x: "".join(x["color_identity"]), axis=1)
df["color_identity"]
```
